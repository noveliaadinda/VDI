```python
import pandas as pd
```

```python
data_kdrama = pd.read_csv('top_100_k_drama_clean.csv')
data_kdrama.head(10)
```

**2. Berdasarkan data yang anda pilih tentukan tujuan, user, fungsi dan tone (jika ada) yang sesuai dengan visualisasi yang akan anda buat**

A. **Tujuan** dari dataset ini adalah menyajikan wawasan tentang pola pelirisan

D. **Tone** disini netral karena penyajian datanya jelas dan netral




**3. Buatlah visualisasi dengan mengunakan library python seperti matplotlib, seaborn atau tools lain**

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame(data_kdrama)
```

```python
# Visualisasi distribusi tahun rilis
plt.figure(figsize=(4, 4))
sns.histplot(df['Year of release'], bins=5, kde=True,color='purple')
plt.title('Distribusi Tahun Rilis K-Drama Top 100')
plt.xlabel('Year of Release')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot: Distribusi rating berdasarkan Stasiun Televisi atau Network
plt.figure(figsize=(4, 4))
sns.boxplot(x='Network', y='Rating', data=df, palette=['pink'])
plt.title('Distribusi Rating Berdasarkan Stasiun Televisi')
plt.xlabel('Network')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
```
