```python
import pandas as pd
```

```python
data_kdrama = pd.read_csv('top_100_k_drama_clean.csv')
data_kdrama.head(10)
```

**2. Berdasarkan data yang anda pilih tentukan tujuan, user, fungsi dan tone (jika ada) yang sesuai dengan visualisasi yang akan anda buat**

A. **Tujuan** dari dataset ini adalah menyajikan wawasan tentang pola pelirisan

C. **Fungsi** akan menvisualisasikan korelasi antar rating, durasi, episode, jumlah episode, serta genre untuk menemukan pola  kdrama yang disukai penonton.

D. **Tone** disini netral karena penyajian datanya jelas dan netral




**3. Buatlah visualisasi dengan mengunakan library python seperti matplotlib, seaborn atau tools lain**

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame(data_kdrama)
```

```python
# Boxplot: Distribusi rating kdrama top 100 berdasarkan Stasiun Televisi atau Network
plt.figure(figsize=(4, 4))
sns.lineplot(x='Network', y='Rating', data=df, color='pink')
plt.title('Distribusi Rating Berdasarkan Stasiun Televisi')
plt.xlabel('Network')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histplot : Distribusi durasi k-drama top 100
plt.figure(figsize=(10, 6))
sns.histplot(df['Duration/min'], bins=15, kde=True, color='purple')
plt.title('Distribusi Durasi K-Drama')
plt.xlabel('Durasi (Menit)')
plt.ylabel('Jumlah K-Drama')
plt.show()
```

```python
# Pie Chart : Proporsi K-drama top 100 berdasarkan tahun rilis
year_counts = df['Year of release'].value_counts()

plt.figure(figsize=(10, 10))
plt.pie(year_counts, labels=year_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('K-Drama Berdasarkan Tahun Rilis')
plt.axis('equal')
plt.show()
```

```python
# Bar Plot : Proporsi K-drama top 100 berdasarkan Number of episode

episode_counts = df['Number of Episode'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.barplot(x=episode_counts.index, y=episode_counts.values, color='pink')
plt.title('Jumlah K-Drama Berdasarkan Jumlah Episode')
plt.xlabel('Jumlah Episode')
plt.ylabel('Jumlah K-Drama')
plt.xticks(rotation=45)
plt.show()
```
